# relacing the Cloud9am & Cloud3pm with mean

```python
rf = rf.drop(['Evaporation','Sunshine'], axis = 1)
rf['Cloud9am'] = rf.Cloud9am.mean()
rf['Cloud3pm'] = rf.Cloud3pm.mean()
rf.dropna(inplace = True)
np.round(rf.isnull().mean())
```

# Let's Analyse the data :

    Let's start with spliting the date colimn into year, month and day

```python
#first convert this 'date' column into datetime type

rf['Date'] = pd.to_datetime(rf['Date'])
```

```python
#split date 

rf['Year'] = rf['Date'].dt.year
rf['Quarter'] = rf['Date'].dt.quarter
rf['Month'] = rf['Date'].dt.month
rf['Weakday'] = rf['Date'].dt.weekday
rf['Day'] = rf['Date'].dt.day
```

```python
#Setting Date column as index
rf = rf.set_index('Date')
```

# before dig deep into time series analysis let's see how rain occurs with respect to Year

```python
sns.set_style('darkgrid')
#sns.set_palette('Set1')
plt.figure(figsize=[15,6])
ax = sns.countplot(x = 'Year', hue = 'RainTomorrow',edgecolor=(0,0,0),linewidth=2,palette="Accent_r", data = rf)
ax.set_title( "Occurence of rain in Years",size = 30 )
plt.show()
```

Here we are seeing a seasonality.From a Certain year Rain starts increasing and after 3-4 years it's reaches to pick and start decreasing.........

```python
sns.set_style('darkgrid')
sns.set_palette('Set2')
plt.figure(figsize=[15,6])
ax = sns.scatterplot(x = 'Month', y = 'Rainfall',data = rf)
ax.set_title( "Amount of  rain in Month",size = 30 )
plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_palette('spring_r')
plt.figure(figsize=[15,6])
ax = sns.countplot(x = 'Day', hue = 'RainTomorrow',edgecolor=(0,0,0),linewidth=2,data = rf)
ax.set_title( "Occurence of rain in day of month",size = 30 )
plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_palette('plasma')
plt.figure(figsize=[15,6])
ax = sns.scatterplot(x = 'Day', y = 'Rainfall',data = rf)
ax.set_title( "Amount of rain in day of month",size = 30 )
plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_palette('plasma_r')
plt.figure(figsize=[15,6])
ax = sns.countplot(x = 'Quarter', hue = 'RainTomorrow',edgecolor=(0,.78,1),linewidth=3,data = rf)
ax.set_title( "Occurence of rain in Quarter",size = 30 )
plt.show()
```
